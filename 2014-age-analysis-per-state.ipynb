```python
df["MINAGE"] = 2014 - df["ADMITYR"] + df["AGEADMIT"]
display(df)
```

```python
grouped_df = df.groupby("STATE")
for state, prisoners in grouped_df:
    print(state)
    min_age = pd.DataFrame({'Total': prisoners.groupby( ["MINAGE"] ).size()}).reset_index()
    #display(min_age)
    not_reported = min_age[min_age["MINAGE"] < 0]
    if not_reported.shape[0] > 0: 
        #print(not_reported)
        print("Not reported " + str(not_reported["Total"].iloc[0]))
        
    min_age = min_age[min_age["MINAGE"] > 0]
    display(min_age)
    #ax = min_age["Total"].plot()
    #ax.set_xticklabels(min_age["MINAGE"].tolist())
    #df.plot(x='col_name_1', y='col_name_2', style='o')

    #min_age["Total"].plot()
    min_age.plot(x="MINAGE", y="Total")
    
    plt.show()
    #display(not_reported)
```
