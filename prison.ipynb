```python
df = df[df["AGEYREND"] == 5]
df = pd.DataFrame({'count' : df.groupby( [ "RPTYEAR", "STATE"] ).size()}).reset_index()

display(df)
df = df.pivot(index = 'RPTYEAR', columns = 'STATE')

display(df)
```

```python
default_dpi = mpl.rcParamsDefault['figure.dpi']
mpl.rcParams['figure.dpi'] = default_dpi*1.5

df.plot()

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          ncol=5)

print("Year-end Prison Population Age 55+")


plt.show()
```
